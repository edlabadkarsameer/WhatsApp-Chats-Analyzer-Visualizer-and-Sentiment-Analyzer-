```python
import numpy
import re
import pandas as pd
import datetime
```

```python
f = open('WhatsApp Chat with AM logo ka group.txt','r',encoding='utf-8')

data =f.read()

pattern ='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
messages = re.split(pattern,data)[1:]

l = len(messages)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message' :messages, 'date': dates})

df
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y, %H:%M - ')
```

```python
df.rename(columns={'message_data': 'date'}, inplace =True)
df.head()
```

```python
df.shape
```

```python
users = []
messages =[]
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['users']= users
df['message'] = messages
df.drop(columns = ['user_message'], inplace=True)
df.head()
        
```

```python
df['year'] = df['date'].dt.year
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df['day'] = df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df['minute'] = df['date'].dt.minute
```

# sentiment analysis

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

text = df.message[1]

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Analyze the sentiment of the text
sentiment = sia.polarity_scores(text)

# Save the sentiment scores in separate variables
neg_score = sentiment['neg']
neu_score = sentiment['neu']
pos_score = sentiment['pos']
compound_score = sentiment['compound']

# Print the sentiment scores
print('Negative score:', neg_score)
print('Neutral score:', neu_score)
print('Positive score:', pos_score)
print('Compound score:', compound_score)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
neg_score = 0
neu_score = 0
pos_score = 0
compound_score = 0
for i in range(1,l-1):
    text = df.message[i]
    sia = SentimentIntensityAnalyzer()
    sentiment = sia.polarity_scores(text)
    neg_score = sentiment['neg']+neg_score
    neu_score = sentiment['neu']+neu_score
    pos_score = sentiment['pos']+pos_score
    compound_score = sentiment['compound']+compound_score
    
neg_score = neg_score/l
neu_score = neu_score/l
pos_score = pos_score/l
compound_score = compound_score/l
print('Negative score:', neg_score)
print('Neutral score:', neu_score)
print('Positive score:', pos_score)
print('Compound score:', compound_score)


max_score = max((neg_score, 'negative'), (neu_score, 'neutral'), (pos_score, 'positive'), key=lambda x: x[0])

# Print the name of the variable with the highest score
print('Average Sentiment:', max_score[1])
```
